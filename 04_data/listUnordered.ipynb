```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
    
    def set(self, data):
        self.data = data
    
    def getVal(self):
        return self.data
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def add(self, node):
        if self.head:
            currentNode = self.head
            while currentNode.next:
                currentNode = currentNode.next
            currentNode.next = node
        else:
            self.head = node
            
    def remove(self, node):
        if self.isEmpty():
            print('List is already empty.')
            return
        
        if self.head == node:
            self.head = self.head.next
        else:
            currentNode = self.head
            while currentNode.next:
                if currentNode.next == node:
                    currentNode.next = currentNode.next.next
                    break
                currentNode = currentNode.next
    
    def visualise(self):
        objs = []
        if self.head:
            currentNode = self.head
            while currentNode:
                objs.append(currentNode.data)
                currentNode = currentNode.next
        print(objs)
        
    def isEmpty(self):
        return self.head == None
    
    def search(self, data):
        result = False
        currentNode = self.head
        while currentNode:
            if currentNode.data == data:
                result = True
                break
            currentNode = currentNode.next
        return result
```

```python
ll = LinkedList()
ll.visualise()

n1 = Node(3)
n2 = Node(6)
n3 = Node(9)

ll.remove(n1)

ll.add(n1)
ll.add(n2)
ll.visualise()

ll.add(n3)
ll.visualise()

ll.remove(n3)
ll.visualise()

ll.search(9)
```
